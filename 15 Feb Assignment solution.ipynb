Multiprocessing refers to the ability of a system to support more than one processor at the same time. Applications in a multiprocessing system are broken to smaller routines that run independently. The operating system allocates these threads to the processors improving performance of the system.

The multiprocessing package offers both local and remote concurrency, effectively side-stepping the Global Interpreter Lock by using subprocesses instead of threads. Due to this, the multiprocessing module allows the programmer to fully leverage multiple processors on a given machine.

# Q3

```python
# importing the multiprocessing module 
import multiprocessing 

def print_cube(num): 
	""" 
	function to print cube of given num 
	"""
	print("Cube: {}".format(num * num * num)) 

def print_square(num): 
	""" 
	function to print square of given num 
	"""
	print("Square: {}".format(num * num)) 

if __name__ == "__main__": 
	# creating processes 
	p1 = multiprocessing.Process(target=print_square, args=(10, )) 
	p2 = multiprocessing.Process(target=print_cube, args=(10, )) 

	# starting process 1 
	p1.start() 
	# starting process 2 
	p2.start() 

	# wait until process 1 is finished 
	p1.join() 
	# wait until process 2 is finished 
	p2.join() 

	# both processes finished 
	print("Done!") 
```

# Q4

* A process pool is a programming pattern for automatically managing a pool of worker processes.

* The pool is responsible for a fixed number of processes.

* It controls when they are created, such as when they are needed.

* When the workload decreases, it can terminate idle processes to free resources.

* This allows you to write concurrent code in an easier way than using threads or multiprocessing directly.

* It also controls what they should do when they are not being used, such as making them wait without consuming computational resources.
* The pool can provide a generic interface for executing ad hoc tasks with a variable number of arguments, much like the target property on the Process object, but does not require that we choose a process to run the task, start the process, or wait for the task to complete.

* Python provides a process pool via the multiprocessing.Pool class.

* The multiprocessing.pool.Pool class can also be accessed by the alias multiprocessing.Pool. They can be used interchangeably.

* It allows tasks to be submitted as functions to the process pool to be executed concurrently.

```python
# https://www.geeksforgeeks.org/synchronization-pooling-processes-python/
# Python program to understand 
# the concept of pool 
import multiprocessing 
import os 

def square(n): 
	print("Worker process id for {0}: {1}".format(n, os.getpid())) 
	return (n*n) 

if __name__ == "__main__": 
	# input list 
	mylist = [1,2,3,4,5] 

	# creating a pool object 
	p = multiprocessing.Pool() 

	# map list to target function 
	result = p.map(square, mylist) 

	print(result) 
```

```python
# importing the multiprocessing module 
import multiprocessing 

def name(nm): 
	print(nm) 

def address(add): 
	print(add) 

def salary(sal): 
	print(sal) 

def age(age): 
	print(age) 

if __name__ == "__main__": 
	# creating processes 
	p1 = multiprocessing.Process(target=name, args=("Rishav", )) 
	p2 = multiprocessing.Process(target=address, args=("Delhi", )) 
	p3 = multiprocessing.Process(target=salary, args=(100000, )) 
	p4 = multiprocessing.Process(target=age, args=(25, )) 

	# starting process 1 
	p1.start() 
	# starting process 2 
	p2.start() 
	p3.start() 
	p4.start() 

	# wait until process 1 is finished 
	p1.join() 
	# wait until process 2 is finished 
	p2.join() 
	p3.join()
	p4.join()

	# both processes finished 
	print("Done!") 
```
