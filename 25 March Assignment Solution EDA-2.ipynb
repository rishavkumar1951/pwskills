### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df.Installs.unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df[cols]=df[cols].str.replace(item,'')
```

```python
df['Installs']=df['Installs'].str.split('+').str[0]
```

```python
df.Installs.unique()
```

```python
## Lets go ahead and explore more data
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
df.drop(df.index[(df['Installs']=='Free')], axis=0, inplace=True)
df.drop(df.index[(df['Installs']=='0')], axis=0, inplace=True)
df.Installs.unique()
```

```python
df['Installs'] =df['Installs'].astype(float)
```

```python
df.Installs.unique()
```

```python
top_10_apps = df.nlargest(10, 'Installs')
```

```python
top_10_apps[['App', 'Installs']]
```

```python
top_10_apps.Installs = top_10_apps.Installs/1000000000# converting into billions
df2 = top_10_apps.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")
ax = sns.barplot(x = 'Installs' , y = 'App' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Apps in Play Store", size = 20)
```

### Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify the most popular app categories. How would you approach this task, and what features would you analyze to make recommendations to the company?

```python
df_cat_installs = df.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")
ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

### Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to launch a new app. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

```python
# I can not find any attribute which can tell us time when we should launch our application.
# If we want then we can use Last Updated columns to identify the best time to launch a new app

df
```

```python
df['Last Updated']=df['Last Updated'].str.split(' ').str[0]
```

```python
df['Last Updated'].unique()
```

```python
df['Last Updated']=df['Last Updated'].map({'January':1,'February':2,'March':3,'April':4,'May':5,'June':6,'July':7,'August':8,'September':9,'October':10,'November':11, 'December':12})
```

```python
df['Last Updated'].unique()
```

```python
df_cat_installs = df.groupby(['Last Updated'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")
ax = sns.barplot(x = 'Last Updated' , y = 'Installs' , data = df2 )
ax.set_xlabel('Month')
ax.set_ylabel('No. of Installations in Billions')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

## We can clearly see that July and August has highest number of installation, so launching app in this month would be best time.
