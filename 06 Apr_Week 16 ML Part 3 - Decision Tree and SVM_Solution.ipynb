### Role of Support Vectors in SVM

Support Vectors are the data points that lie closest to the decision boundary (or hyperplane) in a Support Vector Machine (SVM) algorithm. They are crucial in defining the position and orientation of the hyperplane because the SVM model aims to maximize the margin between the support vectors of different classes. The support vectors are the key elements that the SVM uses to build the optimal hyperplane.

### Key Points about Support Vectors:

1. **Defining the Margin**:
   - The margin is the distance between the hyperplane and the closest data points from both classes. Support vectors are the data points that lie on the edge of this margin.

2. **Influence on the Hyperplane**:
   - Only the support vectors determine the hyperplane. Other data points that are further away from the margin do not affect the position of the hyperplane.

3. **Maximizing the Margin**:
   - SVM aims to find the hyperplane that maximizes the margin between support vectors of different classes, which helps in achieving better generalization on unseen data.

4. **Support Vectors and Slack Variables**:
   - In cases where the data is not linearly separable, SVM introduces slack variables to allow some misclassifications. Support vectors may then lie within the margin or on the wrong side of the hyperplane but with a penalty.

# Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?

To illustrate the concepts of Hyperplane, Marginal plane, Soft margin, and Hard margin in Support Vector Machines (SVM), let's consider a simple 2D example where we have two classes of points that are separable or nearly separable.

### Example Dataset

Let's define two classes in a 2D space:

- Class A (Positive class): Points in the lower-left quadrant.
- Class B (Negative class): Points in the upper-right quadrant.

#### Example Points:
- Class A: (1, 2), (2, 3), (3, 3)
- Class B: (6, 6), (7, 7), (8, 7)

### Graphical Illustration

Let's plot these points and visualize the concepts:

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data points
classA = np.array([[1, 2], [2, 3], [3, 3]])
classB = np.array([[6, 6], [7, 7], [8, 7]])

# Plotting the points
plt.figure(figsize=(8, 6))
plt.scatter(classA[:, 0], classA[:, 1], color='blue', label='Class A (Positive)')
plt.scatter(classB[:, 0], classB[:, 1], color='red', label='Class B (Negative)')
plt.title('Example Dataset for SVM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

This code will plot the points in a 2D space, where Class A (blue) is in the lower-left and Class B (red) is in the upper-right.

### Concepts Visualization

1. **Hyperplane**:
   - The hyperplane is the decision boundary that separates the two classes. In a 2D space, it's a line.

2. **Margins**:
   - Margins are the lines parallel to the hyperplane that pass through the support vectors (points closest to the hyperplane).

3. **Hard Margin vs Soft Margin**:
   - **Hard Margin**: A strict margin where all points are correctly classified with no allowance for misclassifications. It may not always be feasible if the data is not perfectly separable.
   - **Soft Margin**: Allows for some misclassifications (slack variables) to find a margin that separates most points correctly. This is more flexible and practical for real-world datasets.

### Adding Hyperplane and Margins

Let's visualize the hyperplane and margins:

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data points
classA = np.array([[1, 2], [2, 3], [3, 3]])
classB = np.array([[6, 6], [7, 7], [8, 7]])

# Plotting the points
plt.figure(figsize=(8, 6))
plt.scatter(classA[:, 0], classA[:, 1], color='blue', label='Class A (Positive)')
plt.scatter(classB[:, 0], classB[:, 1], color='red', label='Class B (Negative)')
plt.title('Example Dataset for SVM')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot the hyperplane (decision boundary)
plt.plot([3, 8], [8, 3], linestyle='--', color='green', label='Hyperplane')

# Plot the margins (parallel lines to the hyperplane)
plt.plot([2, 9], [9, 2], linestyle=':', color='orange', label='Margin Plane 1')
plt.plot([4, 7], [7, 4], linestyle=':', color='orange', label='Margin Plane 2')

plt.legend()
plt.grid(True)
plt.xlim(0, 10)
plt.ylim(0, 10)

# Show plot
plt.show()
```

### Explanation

- **Hyperplane**: The green dashed line represents the hyperplane that separates the two classes. In this case, it's a line that roughly divides the points.
  
- **Margins**: The orange dotted lines represent the margins. These are lines parallel to the hyperplane and equidistant from it. They pass through the support vectors (points closest to the hyperplane).

### Soft Margin vs Hard Margin

Let's modify the example to illustrate a soft margin SVM, where some points may be allowed to lie within the margin or on the wrong side of the hyperplane:

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data points
classA = np.array([[1, 2], [2, 3], [3, 3]])
classB = np.array([[6, 6], [7, 7], [8, 7]])

# Plotting the points
plt.figure(figsize=(8, 6))
plt.scatter(classA[:, 0], classA[:, 1], color='blue', label='Class A (Positive)')
plt.scatter(classB[:, 0], classB[:, 1], color='red', label='Class B (Negative)')
plt.title('Example Dataset for SVM (Soft Margin)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot the hyperplane (decision boundary)
plt.plot([3, 8], [8, 3], linestyle='--', color='green', label='Hyperplane')

# Plot the margins (parallel lines to the hyperplane)
plt.plot([1, 9], [9, 1], linestyle=':', color='orange', label='Margin Plane 1')
plt.plot([2, 7], [7, 2], linestyle=':', color='orange', label='Margin Plane 2')

# Plot the support vectors
plt.scatter([2, 3, 7], [3, 3, 6], color='green', marker='o', s=150, edgecolors='black', label='Support Vectors')

plt.legend()
plt.grid(True)
plt.xlim(0, 10)
plt.ylim(0, 10)

# Show plot
plt.show()
```

# Q6. SVM Implementation through Iris dataset.
- Load the iris dataset from the scikit-learn library and split it into a training set and a testing set
- Train a linear SVM classifier on the training set and predict the labels for the testing set
- Compute the accuracy of the model on the testing set
- Plot the decision boundaries of the trained model using two of the features
- Try different values of the regularisation parameter C and see how it affects the performance of the model.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import seaborn as sns

from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
dataset.target
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
#Independent and dependent features
X=df.iloc[:,:-1]
y=dataset.target
```

```python
### train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.size, y_train.size, X_test.size, y_test.size
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
## Prediction
y_pred=svc.predict(X_test)
```

```python
y_pred
```

```python
y_pred.size
```

```python
X_test.size
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

## Hyperparametere Tuning With SVC

```python
from sklearn.model_selection import GridSearchCV
 
# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel':['linear']
              }
```

```python
grid.fit(X_train,y_train)  # cross is 5 so 5x5x5=125 combinations will be
```

```python
grid.best_params_
```

```python
## Prediction
y_pred4=grid.predict(X_test)
print(classification_report(y_test,y_pred4))
print(confusion_matrix(y_test,y_pred4))
print(accuracy_score(y_test,y_pred))
```
