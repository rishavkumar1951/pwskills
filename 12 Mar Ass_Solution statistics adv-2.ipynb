### Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

To calculate the 95% confidence interval, we need to know either the sample size or that we are using the sample standard deviation for the calculation. I'll proceed with the assumption that the sample size is known and use the sample size of 30 as an example. Hereâ€™s the Python code to calculate the confidence interval:

```python
import scipy.stats as stats

# Given values
mean = 50
std_dev = 5
confidence_level = 0.95
sample_size = 30  # Assuming a sample size of 30 for the calculation

# Calculate the standard error
standard_error = std_dev / (sample_size ** 0.5)

# Calculate the margin of error
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1) * standard_error

# Calculate the confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
confidence_interval
```

### Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.



To conduct a chi-square goodness of fit test, we need the observed frequencies of each color of M&Ms. If you provide the observed counts, I can use them for the test. If not, I will create a hypothetical dataset for demonstration purposes.

Let's proceed with a hypothetical dataset of observed M&M counts:

- Blue: 25
- Orange: 18
- Green: 22
- Yellow: 12
- Red: 9
- Brown: 24

We'll compare these observed counts against the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown.

Here's the Python code to perform the chi-square goodness of fit test:

```python
import scipy.stats as stats

# Observed frequencies of M&Ms in the sample bag (hypothetical data)
observed_frequencies = [25, 18, 22, 12, 9, 24]  # [blue, orange, green, yellow, red, brown]

# Total number of M&Ms
total_m_and_ms = sum(observed_frequencies)

# Expected frequencies based on the given distribution
expected_proportions = [0.20, 0.20, 0.20, 0.10, 0.10, 0.20]
expected_frequencies = [p * total_m_and_ms for p in expected_proportions]

# Perform the chi-square goodness of fit test
chi_square_stat, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

chi_square_stat, p_value
```

The chi-square goodness of fit test results are as follows:
- **Chi-square statistic**: 1.77
- **p-value**: 0.880

### Interpretation:
With a significance level of 0.05, we compare the p-value to determine whether to reject the null hypothesis. Since the p-value (0.880) is much greater than 0.05, we fail to reject the null hypothesis.

### Conclusion:
There is not enough evidence to conclude that the distribution of colors of M&Ms in the bag differs from the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. The observed frequencies are consistent with the expected distribution.

### Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.



```python
import scipy.stats as stats
import math

# Given values
mean = 75
std_dev = 12
confidence_level = 0.90
sample_size = 30

# Calculate the z-score for the 90% confidence level
z_score = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate the standard error of the mean
standard_error = std_dev / math.sqrt(sample_size)

# Calculate the margin of error
margin_of_error = z_score * standard_error

# Calculate the confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)
print("90% confidence interval:", confidence_interval)
```

### Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Parameters
df = 10  # degrees of freedom
chi_square_statistic = 15

# Create the x values
x = np.linspace(0, 30, 1000)

# Create the y values for the chi-square distribution
y = chi2.pdf(x, df)

# Plot the chi-square distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Chi-Square Distribution df=10')
plt.fill_between(x, y, where=(x >= chi_square_statistic), color='gray', alpha=0.5, label='Chi-Square Statistic = 15')

# Label the axes
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

### Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.





```python
import math

# Given values
n = 1000
p_hat = 520 / 1000  # sample proportion
confidence_level = 0.99

# Z-score for 99% confidence level
z_score = 2.576

# Calculate the standard error
standard_error = math.sqrt((p_hat * (1 - p_hat)) / n)

# Calculate the margin of error
margin_of_error = z_score * standard_error

# Calculate the confidence interval
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)
print("99% confidence interval:", confidence_interval)
```

### Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.






```python
import scipy.stats as stats

# Observed frequencies
observed = [45, 55]

# Expected frequencies for a fair coin
expected = [50, 50]

# Calculate the chi-square statistic and the p-value
chi2_statistic, p_value = stats.chisquare(observed, expected)

# Critical value for chi-square distribution with 1 degree of freedom at 0.05 significance level
critical_value = stats.chi2.ppf(0.95, df=1)

print("Chi-square statistic:", chi2_statistic)
print("p-value:", p_value)
print("Critical value at 0.05 significance level:", critical_value)
```

This code will calculate the chi-square statistic, the p-value, and the critical value. You can then compare the chi-square statistic to the critical value to determine whether to reject the null hypothesis.

**Decision Rule:**
- If the chi-square statistic is greater than the critical value, reject the null hypothesis.
- If the chi-square statistic is less than or equal to the critical value, do not reject the null hypothesis.

**Interpretation:**
- If you reject the null hypothesis, it suggests that the coin is biased towards tails.
- If you do not reject the null hypothesis, it suggests that there is not enough evidence to conclude that the coin is biased.

### Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05






```python
import math
from scipy import stats

# Given values
sample_mean = 72
population_mean = 70
sample_std_dev = 10
sample_size = 30
alpha = 0.05

# Calculate the test statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / math.sqrt(sample_size))

# Degrees of freedom
df = sample_size - 1

# Calculate the critical t-value for a two-tailed test
critical_t = stats.t.ppf(1 - alpha/2, df)

# Calculate the p-value
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))

print("t-statistic:", t_statistic)
print("Critical t-value:", critical_t)
print("p-value:", p_value)
```
